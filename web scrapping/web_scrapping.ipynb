```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
```

```python
#fetching reviews or comments
```

```python
reviews = soup.findAll('div',{'class':'ZmyHeo'})
for review in reviews:
    print(review.get_text()+"\n")
```

```python
#fetching retings
#avg overall ratings
ratings = soup.find('div',{'class':'ipqd2A'});
print(ratings.get_text())
```

```python
#fetching individual ratings
individual_ratings = soup.findAll('div',{'class':'XQDdHH Ga3i8K'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text()+'\n')
```

```python
#fetching tags
tags = soup.find('span',{'class':'_6EBuvT OHdLLG'})
print(tags.get_text())
```

```python
#fetching customer names
customer_names = soup.findAll('p',{'class':'_2NsDsF AwS1CA'})
for customer_name in customer_names:
    print(customer_name.get_text()+'\n')
```
